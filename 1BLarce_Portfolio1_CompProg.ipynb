This is a calculator in determining the mass, acceleration, or force of a given matter.

PROBLEM 1: Calculating Mass (m)
Scenario: You are helping move furniture, and you push a heavy crate with a 
force of 150 N. The crate accelerates across the floor at a rate of 3 m/s^2. 
What is the mass of the crate?

PROBLEM #2: Calculating Acceleration (a)
Scenario: A professional bowler throws a 7 kg bowling ball down the lane. 
They exert a forward force of 35 N on the ball at the moment of release. 
How fast does the ball accelerate?


PROBLEM #3: Calculating Force (F)
Scenario: An engineer is testing a new motorized cart that has a total mass 
of 200kg. The design requirements state that the cart must be able to accelerate 
from a standstill at a rate of 1.5 m/s^2. How much force do it need to produce?

```python
# Computing for Mass 
# This sets up the variable Force and its given value for the value.
Force = float(input("Force (N):")) 

# This sets up the variable Acceleration and its given value for the value.
Acceleration = float(input("Acceleration (m/s^2):")) 

# This section complete the instructions given by the code and executes it.
Mass=float(Force/Acceleration) 
# This displays the result to the user.
print("The mass of the crate is", Mass, "kg.") 
```
